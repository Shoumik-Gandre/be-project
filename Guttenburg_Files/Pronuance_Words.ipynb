```python
targets = ['man', 'woman']
filtered = []
for text in gutenberg.fileids():
    words = gutenberg.words(text)
    for word in words:
        word_lc = word.lower()              # word.lower() converts the word to lower case
        if word_lc in targets:
            filtered.append((word_lc, text))
```

```python
d = {}
for w in new_words:
    x = new_words.count(w)
    if w not in d.keys():
        d.update({w:x})
print(d)
        
```

```python
print(max(d.values()))
```

```python
for i,j in d.items():
    if j==max(d.values()):
        print(str(i)+','+str(j))
```

```python

for i,j in d.items():
    with open("words1.csv","a") as f:
        f.write(str(i)+','+str(j)+"\n")
        f.close()
```
